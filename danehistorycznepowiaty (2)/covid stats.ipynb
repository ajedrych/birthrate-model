```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_rows", 10)
```

```python
df = pd.read_csv('20220107074508_rap_rcb_pow_eksport.csv', delimiter=';', encoding="utf8")
df
```

```python
df = df[df.powiat_miasto == "Warszawa"]
df
```

```python
all_files = glob.glob(os.path.join("D:\Licencjat\dane_historyczne_powiaty_zakazenia_COVID19", "*.csv"))

result= pd.DataFrame()
for f in all_files:
    df = pd.read_csv(f, delimiter=';', encoding='ISO-8859-1')
    result = result.append(df[df.powiat_miasto == "Warszawa"])

result
```

```python
pd.to_datetime(result['stan_rekordu_na'])
result=result.reset_index(drop=True)
result
```

```python
pd.set_option("display.max_rows", 20)
display(result)
```

```python
from datetime import date, timedelta

start_date = date(2020, 12, 23)
end_date = date(2021, 1, 26)
delta = timedelta(days=1)

while start_date <= end_date:
    for j in range (30, 65):
        result.loc[j, 'stan_rekordu_na'] = start_date.strftime("%Y-%m-%d")
        start_date += delta

result
```

```python
result.to_csv (r'D:\Licencjat\warszawa_zakazenia_covid19.csv', index = False, header=True)
```
